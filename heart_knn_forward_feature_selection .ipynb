```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("heart 2.csv")
df_e=pd.get_dummies(df)
```

```python
X=df_e.drop(['HeartDisease'],axis=1)
y=df_e['HeartDisease']
```

```python
from mlxtend.feature_selection import SequentialFeatureSelector as sfs
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
```

```python
neigh = KNeighborsClassifier()
```

```python
sfs1 = sfs(neigh, k_features=15, forward=True, verbose=2, scoring='neg_mean_squared_error')
```

```python
sfs1 = sfs1.fit(X, y)
```

```python
selected_features = list(sfs1.k_feature_names_)
print(selected_features)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
X = df_e[selected_features]
y = df_e['HeartDisease']

scaler=StandardScaler()
X=scaler.fit_transform(X)

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.4,random_state=42)

neigh = KNeighborsClassifier(n_neighbors = 9).fit(X_train,y_train)
yhat = neigh.predict(X_test)
print("Train set Accuracy :", metrics.accuracy_score(y_train, neigh.predict(X_train)))
print("Test set Accuracy :", metrics.accuracy_score(y_test, yhat))
```

```python
from sklearn.metrics import precision_score, recall_score
from sklearn.metrics import f1_score
```

```python
precision_score(y_test, yhat)
```

```python
recall_score(y_test, yhat)
```

```python
f1_score(y_test, yhat)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, yhat))
```
